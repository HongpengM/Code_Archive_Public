```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import numpy as np
import pandas as pd
```

```python
browser = webdriver.Chrome()
```

```python
browser.get('https://quizlet.com/470137394/400-investment-banking-questions-flash-cards/')
```

```python
qaas = browser.find_elements_by_xpath('//div[@class="SetPageTerms-term"]')
```

```python
for qaa in qaas[0:10]:
    print(qaa.text)
```

```python
len(qaas)
```

```python
_qs = []
_as = []
for qaa in qaas:
    try:
        q = qaa.find_element_by_xpath('div/div/div/div/div[@class="SetPageTerm-side SetPageTerm-smallSide"]')
        a = qaa.find_element_by_xpath('div/div/div/div/div[@class="SetPageTerm-side SetPageTerm-largeSide"]')
        print('Question: ', q.text)
        print('Answer:', a.text)
        print('-'* 40)
        _qs.append(q.text)
        _as.append(a.text)
    except Exception:
        continue
```

```python
len(_qs)
```

```python
df = pd.DataFrame(_qs,_as)
```

```python
df.to_excel('400 questions.xlsx')
```
